```
Add two columns High_Low_Speed and High_Low_HP in the same dataset as we worked in class.
```



````python
import pandas as pd
````

````python
data = pd.read_csv("/content/drive/MyDrive/Pokemon.csv")
data.head(10)
````

````python
data.describe()
````

**FUNCTION OF SET SPEED AND ADDING COLUMN HIGH_LOW_SPEED**

````python
speed_mean = data["Speed"].mean()
print(speed_mean)

def set_speed(val):
    if val < speed_mean:
        return "Low Speed"
    elif val == speed_mean:
        return "Neutral Speed"
    else:
        return "High Speed"
````

````python
# "High_Low_Speed"  -> "High Speed" , "Low Speed"
data["High_Low_Speed"] = data["Speed"].apply(set_speed)
data
````

**FUNCTION OF SET HP AND ADDING COLUMN HIGH_LOW_HP**

````python
HP_mean = data["HP"].mean()
print(HP_mean)

def set_HP(val):
    if val < HP_mean:
        return "Low HP"
    elif val == HP_mean:
        return "Neutral HP"
    else:
        return "High HP"
````

````python
# "High_Low_HP"  -> "High HP" , "Low HP"
data["High_Low_HP"] = data["HP"].apply(set_HP)
data
````
